```python
#Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_palette("GnBu_d")
sns.set_style('whitegrid')
```

```python
#Loading raw data
car_data = pd.read_csv("raw_scraped_data.csv")
```

```python
car_data.head(2)
```

```python
#Exporing descriptive statistics of the dataset
car_data.describe(include='all')
```

```python
#Exporing data info
car_data.info()
```

```python
#Drop columns not required & standardise car_title
df = car_data.drop(['car_id', 'car_decsription', 'source', 'url'], axis=1)
df['car_title'] = 'Mercedes-Benz A Class'
df.head()
```

## Data Exploration

```python
# This makes it very easy to spot anomalies, such as outliers
plt.figure(figsize=[11,5])
sns.distplot(df['price_gbp'])
plt.title('Car Price Distribution Plot')
```

```python
# car and mileage Distribution Plot
plt.figure(figsize=[17,5])
plt.subplot(1,3,1)
sns.distplot(df['year'])
plt.title('Car Year Distribution Plot')

plt.subplot(1,3,2)
sns.distplot(df['mileage'])
plt.title('Car Mileage Distribution Plot')

plt.subplot(1,3,3)
sns.distplot(df['engine'])
plt.title('Engine Distribution Plot')
```

```python
#Scatter Plot of independent variables vs dependent variables
plt.figure(figsize=[20,7])
plt.subplot(1,3,1)
plt.title("Price and Year")
sns.scatterplot(x='year',y='price_gbp',data=df)

plt.subplot(1,3,2)
plt.title("Price and Mileage")
sns.scatterplot(x='price_gbp',y='mileage',data=df)

plt.subplot(1,3,3)
sns.scatterplot(y='price_gbp',x='engine',data=df)
plt.title("Price and Engine")
```
